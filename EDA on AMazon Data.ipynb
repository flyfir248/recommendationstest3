```python
import re

# Function to remove non-ASCII characters from deal titles
def clean_title(title):
    return re.sub(r'[^\x00-\x7F]+', '', title)

# Apply the function to the 'deal_title' column
df['deal_title'] = df['deal_title'].apply(clean_title)

# Display the first few rows to verify the changes
print(df['deal_title'].head())
```

## Extract Domain from Deal URL:

```python
from urllib.parse import urlparse

# Function to extract domain from URL
def extract_domain(url):
    parsed_url = urlparse(url)
    return parsed_url.netloc

# Apply the function to the 'deal_url' column
df['domain'] = df['deal_url'].apply(extract_domain)

# Display the extracted domain
print(df['domain'].head())
```

### Extract Numeric Values from Offer Descriptions

```python
import ast
import re

# Function to extract numeric values from text
def extract_numeric(text):
    numbers = re.findall(r'\d+', text)
    return [int(num) for num in numbers]

# Function to safely extract numeric values from offer descriptions
def safe_extract_numeric(offers):
    try:
        # If offers is already a list of dictionaries
        if isinstance(offers, list):
            return [extract_numeric(offer['description']) for offer in offers]
        
        # If offers is a string, safely evaluate it
        offers_list = ast.literal_eval(offers)
        
        # Check if the evaluated result is a list of dictionaries
        if isinstance(offers_list, list) and all(isinstance(offer, dict) for offer in offers_list):
            return [extract_numeric(offer['description']) for offer in offers_list]
        
        # Return an empty list if the structure is not as expected
        return []
    except (ValueError, SyntaxError):
        return []

# Apply the function to the 'top_offers' column
df['offer_values'] = df['top_offers'].apply(safe_extract_numeric)

# Display the extracted numeric values
print(df['offer_values'].head())
```

### Extract Product Features

```python
# Function to extract product features
def extract_features(description):
    if isinstance(description, str):
        return description.split(',')
    else:
        return []

# Apply the function to the 'product_description' column
df['product_features'] = df['product_description'].apply(extract_features)

# Display the extracted product features
print(df['product_features'].head())
```

## Extract Ratings from Reviews

```python
import pandas as pd
import ast

# Function to safely evaluate and extract ratings from reviews
def safe_extract_ratings(reviews):
    if pd.isna(reviews):  # Handle NaN values
        return []
    try:
        reviews_list = ast.literal_eval(reviews)
        if isinstance(reviews_list, list):  # Check if the result is a list
            return [int(review['rating']) for review in reviews_list if 'rating' in review]
        else:
            return []
    except (ValueError, SyntaxError):
        return []

# Apply the function to the 'top_customer_reviews' column
df['customer_ratings'] = df['top_customer_reviews'].apply(safe_extract_ratings)

# Display the extracted customer ratings
print(df['customer_ratings'].head())
```
